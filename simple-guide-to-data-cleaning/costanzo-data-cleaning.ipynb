```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('modified_titanic_data.csv')
```

```python
df.drop(columns="Unnamed: 13", inplace=True)
```

```python
df[df.duplicated(keep=False)]
```

```python
df = df.drop_duplicates(keep="first")
```

```python
df["Sex"] = df["Sex"].map({
    "male": "male",
    "m": "male",
    "m ": "male",
    "M": "male",
    "F": "female",
    "f": "female",
    "female": "female"
})
```

```python
df['Age'] = df['Age'].replace('-', np.NaN)
df['Age'] = df['Age'].replace('na', np.NaN)
df["Age"].head(10) 
```
